```python
#Generador de contrase√±as
import random
import string

def my_generate(length):
    char = string.ascii_letters
    char += string.digits #"0123456789"
    return "".join(random.choice(char) for _ in range(length))

length = int(input("What size: "))

print("Your password: ",my_generate(length))
```
